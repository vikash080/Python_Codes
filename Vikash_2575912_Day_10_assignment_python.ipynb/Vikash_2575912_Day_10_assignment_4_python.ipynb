Define a year as a "Superman year" whose films feature more Superman characters than Batman. How many years in film history have been Superman years?

```python


# Filter for rows where the character is "Superman" or "Batman"
superman_batman_roles = cast[(cast['character'] == 'Superman') | (cast['character'] == 'Batman')]

# Group the data by year and character and count the number of characters for each year and character
character_counts = superman_batman_roles.groupby(['year', 'character']).size().unstack(fill_value=0)

# Filter for years where there are more Superman characters than Batman characters
superman_years = character_counts[character_counts['Superman'] > character_counts['Batman']]

# Count the number of Superman years
num_superman_years = superman_years.shape[0]

print("Number of years in film history that have been Superman years:", num_superman_years)
```

Plot the number of actor roles each year and the number of actress roles each year over the history of film.

```python
import pandas as pd
import matplotlib.pyplot as plt


cast = pd.read_csv('cast.csv')

# Filter for rows where the type is "actor" or "actress"
actor_actress_roles = cast[(cast['type'] == 'actor') | (cast['type'] == 'actress')]

# Group the data by year and type and count the number of roles for each year and type
roles_by_year_type = actor_actress_roles.groupby(['year', 'type']).size().unstack(fill_value=0)

# Create a line plot to visualize the number of actor and actress roles each year
plt.figure(figsize=(12, 6))
roles_by_year_type['actor'].plot(label='Actor', color='blue')
roles_by_year_type['actress'].plot(label='Actress', color='red')
plt.title('Number of Actor and Actress Roles Each Year')
plt.xlabel('Year')
plt.ylabel('Number of Roles')
plt.legend()
plt.grid(True)
plt.show()
```

Plot the number of actor roles each year and the number of actress roles each year, but this time as a kind='area' plot.

```python

import matplotlib.pyplot as plt

cast = pd.read_csv('cast.csv')

# Filter for rows where the type is "actor" or "actress"
actor_actress_roles = cast[(cast['type'] == 'actor') | (cast['type'] == 'actress')]

# Group the data by year and type and count the number of roles for each year and type
roles_by_year_type = actor_actress_roles.groupby(['year', 'type']).size().unstack(fill_value=0)

# Create an area plot to visualize the number of actor and actress roles each year
plt.figure(figsize=(12, 6))
roles_by_year_type.plot(kind='area', stacked=True)
plt.title('Number of Actor and Actress Roles Each Year')
plt.xlabel('Year')
plt.ylabel('Number of Roles')
plt.legend(['Actor', 'Actress'])
plt.grid(True)
plt.show()
```

Plot the fraction of roles that have been 'actor' roles each year in the history of film.

```python


# Group the data by year and type, and count the number of roles for each year and type
roles_by_year_type = cast.groupby(['year', 'type']).size().unstack(fill_value=0)

# Calculate the fraction of 'actor' roles each year
fraction_actor_roles = roles_by_year_type['actor'] / (roles_by_year_type['actor'] + roles_by_year_type['actress'])

# Create a line plot to visualize the fraction over the years
plt.figure(figsize=(12, 6))
fraction_actor_roles.plot()
plt.title('Fraction of Roles That Are "Actor" Roles Each Year')
plt.xlabel('Year')
plt.ylabel('Fraction')
plt.grid(True)
plt.show()
```

```python




# Filter for rows where the 'n' column is 2 and the 'type' column is 'actor'
supporting_actor_roles = cast[(cast['n'] == 2) & (cast['type'] == 'actor')]


supporting_actor_counts = supporting_actor_roles.groupby('year').size()

# Filter for rows where the 'n' column is 2
supporting_roles = cast[cast['n'] == 2]

# Group the data by year and count the total number of supporting roles each year
total_supporting_counts = supporting_roles.groupby('year').size()


fraction_supporting_actor = supporting_actor_counts / total_supporting_counts

# Create a line plot to visualize the fraction over the years
plt.figure(figsize=(10, 6))
fraction_supporting_actor.plot()
plt.title('Fraction of Supporting Roles (n=2) That Are "Actor" Roles Each Year')
plt.xlabel('Year')
plt.ylabel('Fraction')
plt.grid(True)
plt.show()
```
