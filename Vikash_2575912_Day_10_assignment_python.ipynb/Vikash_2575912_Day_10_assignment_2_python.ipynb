Which three years of the 1930s saw the most films released

```python
import pandas as pd

# Filter for years in the 1930s (1930 to 1939)
films_1930s = titles[(titles['year'] >= 1930) & (titles['year'] <= 1939)]

# Count the number of films released in each year
film_counts_by_year = films_1930s['year'].value_counts()

# Select the top three years with the most films
top_three_years = film_counts_by_year.head(3)

print("Three years in the 1930s with the most films released:")
print(top_three_years)
```

plot the number of films that have been released each decade over the history of cinema.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the titles DataFrame
titles = pd.read_csv('titles.csv')

# Extract the decade from the 'year' column
titles['decade'] = (titles['year'] // 10) * 10

# Group the data by decade and count the number of films
film_counts_by_decade = titles['decade'].value_counts().sort_index()

# Create a bar chart to visualize the data
plt.figure(figsize=(12, 6))
film_counts_by_decade.plot(kind='bar', color='skyblue')
plt.title('Number of Films Released Each Decade')
plt.xlabel('Decade')
plt.ylabel('Number of Films')
plt.xticks(rotation=45)
plt.show()
```

Plot how many roles Mammootty has played in each year of his career.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the cast DataFrame
cast = pd.read_csv('cast.csv')

# Filter for rows where the name is "Mammootty"
mammootty_roles = cast[cast['name'] == 'Mammootty']

# Group the data by year and count the number of roles each year
roles_by_year = mammootty_roles.groupby('year').size()

# Create a line plot to visualize the data
plt.figure(figsize=(12, 6))
roles_by_year.plot(kind='line', color='skyblue')
plt.title('Number of Roles Played by Mammootty Each Year')
plt.xlabel('Year')
plt.ylabel('Number of Roles')
plt.show()
```

What are the 10 most frequent roles that start with the phrase "Patron in"?

How many leading (n=1) roles were available to actors, and how many to actresses, in the 1950s?

```python


# Filter for rows where the year is in the 1950s
roles_1950s = cast[(cast['year'] >= 1950) & (cast['year'] <= 1959)]

# Count the number of leading roles (n=1) for actors and actresses separately
leading_roles_actors = roles_1950s[(roles_1950s['n'] == 1) & (roles_1950s['type'] == 'actor')]
leading_roles_actresses = roles_1950s[(roles_1950s['n'] == 1) & (roles_1950s['type'] == 'actress')]

# Get the counts for both actors and actresses
count_leading_roles_actors = leading_roles_actors.shape[0]
count_leading_roles_actresses = leading_roles_actresses.shape[0]

print("Number of leading roles (n=1) available to actors in the 1950s:", count_leading_roles_actors)
print("Number of leading roles (n=1) available to actresses in the 1950s:", count_leading_roles_actresses)
```
